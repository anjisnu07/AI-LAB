```python
import numpy as np
import matplotlib.pyplot as plt


def function1(x):
    return x**2 - 4*x

def function2(x):
    return 4*x**2 + 5*x + 2

def function3(x):
    return np.sin(3*x) + np.sin(2*x) + np.sin(x)


```

```python
# Gradient descent function
def gradient_descent(func, initial_point, learning_rate, num_iterations):
    x = initial_point
    x_history = [x]
    for _ in range(num_iterations):
        gradient = 2*x - 4
        x -= learning_rate * gradient
        x_history.append(x)
    return x, x_history

# Define parameters
initial_points = [10] * 9
learning_rates = [0.01, 0.1, 1.0, 0.5, 0.75]
num_iterations = 100

# Plotting function
def plot_results(func, initial_points, learning_rates, num_iterations):
    x_values = np.linspace(-4, 6, 100)
    y_values = func(x_values)
    plt.figure(figsize=(12, 8))
    plt.plot(x_values, y_values, label="Function")
    for i, (initial_point, learning_rate) in enumerate(zip(initial_points, learning_rates), start=1):
        minima, minima_history = gradient_descent(func, initial_point, learning_rate, num_iterations)
        plt.scatter(minima, func(minima), color='red', label=f"Minima {i}")
        plt.plot(minima_history, [func(x) for x in minima_history], label=f"Gradient Descent {i}")
    plt.title("Gradient Descent Optimization")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python

# Plot results for function 1
plot_results(function1, initial_points, learning_rates, num_iterations)

# Plot results for function 2
plot_results(function2, initial_points, learning_rates, num_iterations)

```

```python
initial_points = [0, 1, 1]  # Separate initial points for each combination
learning_rates = [0.01, 0.01, 0.1]
num_iterations = 500

for i, (start_point, learning_rate) in enumerate(zip(initial_points, learning_rates), start=1):
    minima, minima_history = gradient_descent(function3, start_point, learning_rate, num_iterations)
    plt.plot(minima_history, label=f"Start Point={start_point}, Learning Rate={learning_rate}")
plt.title("Gradient Descent Optimization for function3")
plt.xlabel("Iteration")
plt.ylabel("x")
plt.legend()
plt.grid(True)
plt.show()
```
