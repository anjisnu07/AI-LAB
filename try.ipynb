```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func1(x):
    return x**2-4*x
```

```python
def func2(x):
    return 4*x**2 + 5*x + 2
```

```python
def gradient_descent(func,initial_pt,learning_rate,num_iteration):
    x=initial_pt
    x_history=[x]
    for _ in range(num_iteration):
        derivative=2*x-4
        x=x-learning_rate*derivative
        x_history.append(x)
    return x,x_history
```

```python
plt.figure(figsize=(16,12))
x_values=np.linspace(-4,6,100)
plt.plot(x_values,func1(x_values),label="Function")
```

```python
initial_point = 10
learning_rates = [0.01, 0.1, 1.0, 0.5, 0.75]
num_iterations = [500, 100, 100, 100, 100]

for i,(learning_rate,num_iteration) in enumerate(zip(learning_rates,num_iterations),start=1):
    minima,minima_history= gradient_descent(func1,initial_point,learning_rate,num_iteration)
    plt.scatter(minima,func1(minima),label=f"Minima:{i}")
    plt.plot(minima_history,[func1(x) for x in minima_history],label=f"Intermediate step:{i}")
plt.title("gradient_descent for function 1")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def gradient_descent_func2(func,initial_pt,learning_rate,iteration):
    x=initial_pt
    x_history=[x]
    for _ in range(iteration):
        dervative=8*x+5
        x=x-learning_rate*dervative
        x_history.append(x)
    return x,x_history
```

```python
plt.figure(figsize=(12,8))
plt.plot(x_values,func2(x_values),label="Function")
```

```python
for i,(num_iteration,learning_rate) in enumerate(zip(num_iterations,learning_rates),start=1):
    minima,minima_history=gradient_descent_func2(func2,initial_point,learning_rate,num_iteration)
    plt.scatter(minima,func2(minima),label=f"minima {i}")
    plt.plot(minima_history,[func2(x) for x in minima_history],label=f"gradient_descent {i}")
plt.title("gradient_descent for func2")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def func3(x):
    return np.sin(3*x) + np.sin(2*x) + np.sin(x)
```

```python
x_values=np.linspace(-4,4,101)
plt.plot(x_values,func3(x_values),label="function")
```

```python
def gradient_descent_fuc3(func3,initial_pt,learning_rate,iteration):
    x=initial_pt
    x_history=[x]
    for _ in range(iteration):
        gradient=3*np.cos(3*x)+2*np.cos(2*x)+np.cos(x)
        x=x-gradient*learning_rate
        x_history.append(x)
    return x,x_history
```

```python
learning_rates=[0.01,0.01,0.1]
initial_points=[0,1,1]
num_iteration=500
for i,(learning_rate,initial_point) in enumerate(zip(learning_rates,initial_points),start=1):
    minima,minima_history=gradient_descent_fuc3(func3,initial_point,learning_rate,num_iteration)
    plt.scatter(minima,func3(minima),label=f"Minima {i}")
    plt.plot(minima_history,[func3(k) for k in minima_history],label=f"gradient descent {i}")
plt.title("Gradient descent for funct 3")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
