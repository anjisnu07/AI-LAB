```python
def issafe(board,row,col,n):
    for c in range(col):
        if(board[row][c]==1):
            return False
    for i,j in zip(range(row,-1,-1),range(col,-1,-1)):
        if(board[i][j]==1):
            return False
    for i,j in zip(range(row,n,1),range(col,-1,-1)):
        if(board[i][j]==1):
            return False
    return True
```

```python
def n_queen_solve(board,col,solutions,n):
    if(col==n):
        solutions.append([col[:] for col in board])
        return
    for i in range(n):
        if(issafe(board,i,col,n)):
            board[i][col]=1
            n_queen_solve(board,col+1,solutions,n)
            board[i][col]=0
```

```python
def n_queen(n):
    board=[[0]*n for _ in range(n)]
    solutions=[]
    count=1
    n_queen_solve(board,0,solutions,n)
    if not solutions:
        print("not exist")
        return
    else:
        for sol in solutions:
            print(f"solution {count}")
            count=count+1
            for i in sol:
                for cell in i:
                    if(cell==1):
                        print("Q",end=" ")
                    else:
                        print(".",end=" ")
                print()
                    
```

```python
lis=[4,5,6]

for li in lis:
    print(f"For {li} no of queens solutions are:")
    n_queen(li)
```
