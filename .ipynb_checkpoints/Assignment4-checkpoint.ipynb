```python
import matplotlib.pyplot as plt

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
```

```python
class Line:
    def __init__(self,st,en):
        self.st=st
        self.en=en
    def display(self):
        plt.plot([self.st.x,self.en.x],[self.st.y,self.en.y])

```

```python
class FuzzySet:
    def __init__(self, left_bottom, left_top, right_top, right_bottom):
        self.left_bottom = left_bottom
        self.left_top = left_top
        self.right_top = right_top
        self.right_bottom = right_bottom

    def display(self):
        line1 = Line(self.left_bottom, self.left_top)
        line2 = Line(self.left_top, self.right_top)
        line3 = Line(self.right_top, self.right_bottom)
        line1.display()
        line2.display()
        line3.display()
```

```python
line1 = Line(Point(0, 4), Point(4, 0))
line2 = Line(Point(0, 0), Point(4, 4))
plt.figure()
line1.display()
line2.display()
plt.title("Lines")

plt.show()

# Displaying fuzzy set
fs = FuzzySet(Point(20, 0), Point(40, 1), Point(50, 1), Point(70, 0))
plt.figure()
fs.display()
plt.title("Fuzzy Set")
plt.show()
```

```python
def comp(fs):
    left_bottom=Point(fs.left_bottom.x,1-fs.left_bottom.y)
    left_top=Point(fs.left_top.x,1-fs.left_top.y)
    right_top=Point(fs.right_top.x,1-fs.right_top.y)
    right_bottom=Point(fs.right_bottom.x,1-fs.right_bottom.y)

    return FuzzySet(left_bottom,left_top,right_top,right_bottom)
```

```python
c_fs=comp(fs)
c_fs.display()
```

```python
def membership(fs,x):
    if x<=fs.left_bottom.x or x>=fs.right_bottom.x:
        return 0
    elif fs.left_bottom.x<=x<fs.left_top.x:
        return (x-fs.left_bottom.x)/(fs.left_top.x-fs.left_bottom.x)
    elif fs.left_top.x<= x <=fs.right_top.x:
        return 1
    elif fs.right_top.x<x<=fs.right_bottom.x:
        return (fs.right_bottom.x-x)/(fs.right_bottom.x-fs.right_top.x)
    else:
        return 0
```

```python
membership(fs, 69)
```
