```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
dataset
```

```python
import pandas as pd
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
df
```

```python
# No of datapoint
df.shape[0]
```

```python
df['species']=dataset.target_names[dataset.target]
```

```python
df
```

```python
print("b. No. of classes:", df['species'].unique().shape[0])
```

```python
print("c. Data Distribution:")
for species in df['species'].unique():
    count = df[df['species'] == species].shape[0]
    print(f"- {species}: {count}")
```

```python
# No of features
df.shape[1]-1
```

```python
for feature in df.columns[:-1]:
    min=df[feature].min()
    max=df[feature].max()

    print(f"{feature}-min value:{min},max-value:{max}")
```

```python
for feature in df.columns[:-1]:
    mean=df[feature].mean()
    var=df[feature].var()

    print(f"{feature}-Mean value:{mean},varraince-value:{var}")
```

```python
print("g. Classwise min_max of each feature:")
for species in df['species'].unique():
    print(f"\n- {species}:")
    for feature in df.columns[:-1]:
        min_value = df[df['species'] == species][feature].min()
        max_value = df[df['species'] == species][feature].max()
        print(f"  - {feature}: {min_value:.2f} : {max_value:.2f}")
```

```python
print("h. Classwise mean and variance of each feature:")
for species in df['species'].unique():
    print(f"\n- {species}:")
    for feature in df.columns[:-1]:
        mean_value = df[df['species'] == species][feature].mean()
        variance_value = df[df['species'] == species][feature].var()
        print(f"  - {feature}: {mean_value:.2f} : {variance_value:.2f}")
```

```python
import  matplotlib.pyplot as plt
colors = ['blue', 'green', 'red']
for i, feature in enumerate(df.columns[:-1]):
    plt.figure(i + 1)
    for species, color in zip(df['species'].unique(), colors):
        plt.scatter(df[df['species'] == species].index, df[df['species'] == species][feature], c=color, label=species)
    plt.xlabel('Data Point Index')
    plt.ylabel(feature)
    plt.title(f"{feature} Values by Species")
    plt.legend()
    plt.show()
```
